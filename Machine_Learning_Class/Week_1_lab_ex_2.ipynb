When we look at the data, however, we notice that there are certain datapoints that are missing. There are several methods to deal with this. Key resource is here on how to go about it https://towardsdatascience.com/8-methods-for-handling-missing-values-with-python-pandas-842544cdf891
The point is, we could delete the rows that have the missing data
We could set a threshold and say if a row has more than three missing data, then we need to get rid of it
We could also replace all the missing data with 0
But in this case, we are dealing with time series so the best is either to replace with previous value, next value or average. Why? This is temperature data

```python
#Method one, replacing all the datapoints with 0
new_df = new_df.fillna(0) # This replaces all the NaN values with 0s. Is this the best statistical approach though? Worth checking with the lecturers on
new_df
#When run, all the datapoints that were NaN are now 0s
new_df[new_df.isna().any(axis = 1)] #When this is run, we realise that all of them have been removed. However, it is not the most optimal approach
```

```python
#Method 2
df_2.fillna(method = "bfill", limit = 1) #bfill fills the missing values backwards such that the next value replace the NaN value. The limit = 1 ensures that all the it is only the next one-step value that replaces the previous one. This is usefull where a column has several NaNs and there is fear that if we don't have a limit, we will end up filling the entire column with the same data
df_2[df_2.isna().any(axis = 1)] # When run, this returns 1005 as usual hence no filtration has happened yet
```

```python
df_2.info()
#When run , we are told that on the ob_end column, we have 38842 non-null objects which means all of them are clean
#However, max_air_temp has 38343 non-null objects meaning the rest are null
#Additionally, the min_air_temp has 38005 non-null objects which makes it the column with the largest null values
#If you think about the fact that we have 38000 datapoints, getting rid of 1000 would not make so much of a difference. In fact, this would be the easiest solution. But we can find alternative ways.
```

```python
#Method 3 - replace all the missing values with the average value
#The previous method seems to bring up a bit of problem (I think there is some missing knowledge points but we will get it over time)
df_2["max_air_temp"].fillna(df_2.groupby("ob_end_time")["max_air_temp"].transform("median"), inplace = True)
df_2["min_air_temp"].fillna(df_2.groupby("ob_end_time")["min_air_temp"].transform("median"), inplace = True)
df_2.info()
#This still does not work so we will have to go with the easiest alternative - to delete the ones that contains the null values.
```
